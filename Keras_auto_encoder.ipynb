```python
from keras.datasets import mnist
from keras.layers import Input, Dense
from keras.models import Model
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.io
import numpy
```

```python
mat = scipy.io.loadmat('dca_res.mat')
matrix = mat['trainZ1']
feat = matrix.T

mat1 = scipy.io.loadmat('labels.mat')
matrix1 = mat1['labels']

labl = numpy.array(matrix1).reshape(1269,1)
labl = labl

data, target = feat,labl
```

```python
idx = np.random.rand(data.shape[0]) < 0.8
X_train, Y_train = data[idx], target[idx]
X_test, Y_test = data[~idx], target[~idx]
```

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train = X_train.reshape(len(X_train), np.prod(X_train.shape[1:]))
X_test = X_test.reshape(len(X_test), np.prod(X_test.shape[1:]))
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
input_img= Input(shape=(4,))
```

```python
encoded = Dense(units=500, activation='relu')(input_img)
encoded = Dense(units=300, activation='relu')(encoded)
decoded = Dense(units=300, activation='relu')(encoded)
decoded = Dense(units=500, activation='relu')(decoded)
decoded = Dense(units=4, activation='sigmoid')(decoded)
```

```python
autoencoder=Model(input_img, decoded)
```

```python
encoder = Model(input_img, encoded)
```
